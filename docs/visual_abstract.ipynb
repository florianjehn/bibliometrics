```python
import networkx as nx
import matplotlib.pyplot as plt
import random
seed = 5
random.seed(seed)

# Generate the same main component as before
main_component = nx.Graph()
n_main_groups = 6
group_size = 3
group_connection_prob = 0.8
inter_group_connection_prob = 0.05
node_counter = 0
groups = []

for i in range(n_main_groups):
    group_nodes = range(node_counter, node_counter + group_size)
    main_component.add_nodes_from(group_nodes)
    for u in group_nodes:
        for v in group_nodes:
            if u != v and random.random() < group_connection_prob:
                main_component.add_edge(u, v)
    groups.append(list(group_nodes))
    node_counter += group_size

for i in range(n_main_groups):
    for j in range(i + 1, n_main_groups):
        for u in groups[i]:
            for v in groups[j]:
                if random.random() < inter_group_connection_prob:
                    main_component.add_edge(u, v)

# Combine all components (for this example, just use the main component)
G = main_component  # Only use the main connected component

# Visualize the main connected component
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=4)  # Layout for visualization

# Color nodes in the main component's groups distinctly
colors = plt.cm.tab10.colors
for i, group in enumerate(groups):
    nx.draw_networkx_nodes(G, pos, nodelist=group, node_color="grey")


# Remove the axis
plt.axis("off")

plt.savefig(f"docs{os.sep}documents.png")
```
